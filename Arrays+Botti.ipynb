```python
import numpy as np

def Operacion (Accion,ArrayRandom):

    if Accion == 'S':
        Suma=ArrayRandom.sum()
        return(Suma)
    elif Accion =='P':
        Promedio=ArrayRandom.mean()
        return(Promedio)
    elif Accion =='M':
        ValorMax=ArrayRandom.max()
        return(ValorMax)
    elif Accion =='ME':
        Mediana=np.median(ArrayRandom)
        return(Mediana)
    elif Accion =='D':
        StdDev=np.std(ArrayRandom)
        return(StdDev)
    elif Accion =='V':
        Varianza=np.var(ArrayRandom)
        return(Varianza)

Acciones={
                'S':'Suma',
                'P':'Promedio',
                'M':'Maximo',
                'ME':'Mediana',
                'D':'Desvio Std',
                'V':'Varianza',
                'E':'Exit Program'
                }

Exit=False
while Exit == False:
    MaxValue=int(input('Ingrese el valor maximo que puede adoptar cada elemento dentro del Array: '))
    ArrayDim=int(input('Ingrese la dimension del Array: '))
    ArrayRandom=np.random.randint(MaxValue,size=ArrayDim)
    Accion=input('Ingrese la operacion deseada: (S)uma, (P)romedio, (M)aximo, (ME)diana, (D)esvio Std, (V)arianza, (E)xit Program :').upper()
    if Accion not in Acciones.keys():
        print('La opcion seleccionada no es valida.')
    elif Accion =='E':
        Exit=True
    else:
        print(f"El Array aleatorio es {ArrayRandom}: ")
        print(f"El resultado de la operacion {Acciones[Accion]} es {Operacion(Accion,ArrayRandom)}.")3
```
